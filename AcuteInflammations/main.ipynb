```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = open('diagnosis.data','rb')
reader = file.read()
content = reader.decode('utf-16')
lines = content.split('\r\n')
rows = []
for line in lines:
    if line == '': continue
    line = line.replace('\r\n', '')
    fields = line.split('\t')
    row = []
    j = 0
    for field in fields:
        value = float(field.replace(',','.')) if ',' in field else 1. if 'yes' in field else 0.
        row.append(value)
        j += 1
    rows.append(row)
matrix = np.array(rows, dtype = np.float32)
matrix
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense

X_train, X_test, y_train, y_test =  train_test_split(matrix[:,:6], matrix[:,6:], test_size=0.3)
X_testR = X_test
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(6 , activation='relu'))
model.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 4, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 2, kernel_initializer = 'uniform', activation = 'sigmoid'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, batch_size = 10, epochs = 100)
model.save('resources/arem.h5')
```

```python
#Temperature of patient, Occurrence of nausea, Lumbar pain, Urine pushing, Micturition pains, Burning of urethra
# decisions : Inflammation of urinary bladder, Nephritis of renal pelvis origin 
```

```python
preds = model.predict(np.array([X_test[0],X_test[1],X_test[2]]))
preds = np.where(preds > 0.5, "oui", "non")
i=0
for pred in preds:
    print(X_testR[i])
    print("desicion : Inflammation of urinary bladder => " +str(pred[0]), end="\n")
    print("desicion : Nephritis of renal pelvis origin => " +str(pred[1]), end="\n")
    i=i+1
```
