```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pickle
import numpy as np
```

```python
data=pd.read_csv("dataset.csv",sep=",")
data
```

```python
X = data.drop('year', axis=1)
X=data.iloc[:,:12].values
Y=data.iloc[:,13:].values
```

```python
X
```

```python
Y
```

```python
Y = Y.astype(str)
Y = np.char.strip(Y, chars=None)
```

```python
Y = np.where(Y == 'fire', 0, 1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout

```

```python
X = X.astype(float)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
y_train
```

```python
model = Sequential()
model.add(Dense(12, input_shape=(12,), activation='relu'))
model.add(Dense(units = 42, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 126, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dropout(0.3))
model.add(Dense(units = 32, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train.ravel(), batch_size = 4, epochs = 100, validation_split=0.2)
model.save('resources/algerie.h5')
```

```python
import joblib
joblib.dump(sc, 'resources/scaler.pkl')
```

```python
import numpy as np
from keras.models import load_model
import joblib


sc = joblib.load('resources/scaler.pkl')
model = load_model('resources/algerie.h5')

#day	month	Temperature	RH	Ws	Rain	FFMC	DMC	DC	ISI	BUI	FWI
data_pred = sc.transform(np.array([[23,8,56,6,14,0,9,9.4,115.6,7.5,36,15.2],[1, 8, 120,71,14,6.5,64.5,3.3,90.1,1,30.5,0.4],[6,8,31,67,14,0,82.6,5.8,22.2,3.1,7,2.5]]))
pred = model.predict(data_pred)
pred = np.where(pred > 0.5, "Fire", "No Fire")
print(pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=2)

knn.fit(X_train, y_train.ravel())

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```
